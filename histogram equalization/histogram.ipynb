## Importing Library

```python
import cv2
import matplotlib.pyplot as plt
```

## Reading the Image

```python
img = cv2.imread('image.jpg')
imgGray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img)
```

## Histogram Equalization Opertaion

```python
dst =cv2.equalizeHist(imgGray)
dst =cv2.cvtColor(dst,cv2.COLOR_GRAY2BGR)
plt.imshow(dst)
```
